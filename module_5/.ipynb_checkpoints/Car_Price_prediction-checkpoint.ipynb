# Parsing
Приступим к сбору данных на которых будет обучаться модель.
Данные будем собирать с информационного портала https://auto.ru/. Данные соответствуют запросу "Легковые автомобили с пробегом в Москве"

Перед созданием запроса для парсинга отбираем признаки:
- Признаки *equipment_dict* и *complectation_dict* содержат дополнительную информацию об автомобиле и комплектации, но в ряде записей значений таких признаков нет. Можно предположить, что по данным автомобилям просто нет данных или они не заполнены, но не отсутствуют (конечно часть значений признаков может отсутствовать). Возможно в дальнейшем попробуем поработать с отдельными значениями данных признаков, но пока в виду отсутствия более 80% значений признаки *complectation_dict* и *equipment_dict* (так как его заполнение аналогично *complectation_dict*) разбираться не будут
- Признак *image* содержит информацию о ссылке на изображение, так как изображение есть во всех объявлениях признак не информативен, возможно отдельно можно было бы посмотреть влияет ли количество изображений в объявлении, но это уже другая история (если время останется)
- Признак *model_info* содержит информацию о модели, в принципе для исследования модели уже есть признаки *model_name* и *name*
- Признак *parsing_unixtime* содержит информацию о дате парсинга, так как это внешний признак, на который вляет человек собравший данные, добавлять признак не будем, отдельно для каждого объявления будет собран признак с датой публикации объявления
- Признак *super_gen* также содержит информацию уже присутствующую в отдельных признаках
- Признак *vendor* содержит 2 уникальных значения (не смогла разобрать как выбрать, поэтому в виду ограничения по времени признак отсутствует)

# CatBoost

```python
X = data_train.drop(['price'], axis=1,)
y = data_train.price.values
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=VAL_SIZE, shuffle=True, random_state=RANDOM_SEED)
```

CatBoost умеет самостоятельно обрабатывать категориальные признаки, но для корректной обработки нужно их указать:

```python
# чтобы не писать весь список этих признаков, просто вывел их через nunique(). и так сойдет)
X_train.nunique()
```

```python
# Keep list of all categorical features in dataset to specify this for CatBoost
cat_features_ids = np.where(X_train.apply(pd.Series.nunique) < 3000)[0].tolist()
cat_features_ids
```

# Fit

```python
model = CatBoostRegressor(iterations=ITERATIONS,
                          learning_rate=LR,
                          random_seed=RANDOM_SEED,
                          eval_metric='MAPE',
                          custom_metric=['R2', 'MAE']
                          )
model.fit(X_train, y_train,
          cat_features=cat_features_ids,
          eval_set=(X_test, y_test),
          verbose_eval=100,
          use_best_model=True,
          plot=True
          )
```

```python
# оцениваем точность
predict = model.predict(X_test)
print(f"Точность модели по метрике MAPE: {(mape(y_test, predict))*100:0.2f}%")
```

```python
# выведем самые важные признаки для модели
plt.rcParams['figure.figsize'] = (10,10)
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')
```
